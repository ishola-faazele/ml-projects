# VISUALIZE GDP VS LIFE_EXPECTENCY

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("all_data.csv")
```

```python
data.rename(columns = {"Country":"country","Year":'year',"Life expectancy at birth (years)":"life_expectency","GDP":'gdp'}, inplace = True)
```

```python
data.duplicated().sum()
countries = data['country'].unique()
num_rows = (len(countries) // 2) + (len(countries) % 2)
num_rows
```

```python
# Assuming df is your DataFrame

countries = data['country'].unique()
plt.figure(figsize=(20, 20))

for index, country in enumerate(countries):
    country_data = data[data['country'] == country][['gdp', 'life_expectency']]
    plt.subplot(3, 2, index + 1)
    
    # Use a unique color for each country
    color = plt.cm.get_cmap('tab10')(index / len(countries))
    
    plt.scatter(country_data["life_expectency"],country_data["gdp"], label=country, color=color)
    plt.title(country.upper())  # Add a title to each subplot
    plt.xlabel("Life Expectancy")
    plt.ylabel("GDP")
    plt.tight_layout()  

# Add a common legend for all subplots
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.savefig("gdp_vs_life-exptency.png")
plt.show()
```
